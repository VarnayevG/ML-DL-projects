```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None, 'display.max_rows', None)
```

```python
data = pd.read_csv('student-mat.csv', sep=';')
```

```python
num_cols = ['age', 'Medu', 'Fedu', 'traveltime', 'studytime', 'failures', 'famrel', 
            'freetime', 'goout', 'Dalc', 'Walc', 'health', 'absences']
cat_cols = ['school', 'sex', 'address', 'famsize', 'Pstatus', 'Mjob', 'Fjob', 'reason', 'guardian', 
           'schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet', 'romantic',]
target_cols = cols[-3:]
#data = data[num_cols+cat_cols+target_cols]
#cat_cols = data.columns[data.dtypes==object].tolist()
```

```python
data.corr().style.background_gradient(cmap='coolwarm').set_precision(2)
```

Работа с категоральными переменными - one hot encoding

```python
for col in cat_cols:
    n = len(data[col].unique())
    if (n == 2):
        data[col] = pd.get_dummies(data[col], drop_first=True)
```

```python
data = pd.get_dummies(data)
```

Формируем матрицы признаков и матрицы меток:

```python
#все возможные признаки
new_feature_cols = []
for col in data.columns:
    if col != 'G1' and col != 'G2' and col != 'G3':
        new_feature_cols.append(col)
#основные
new_feature_cols = ['studytime', 'failures', 'absences', 'G1', 'G2']
feature_matrix = data[new_feature_cols].values
G1_labels = data['G1'].values
#G1_labels = np.expand_dims(G1_labels, axis=1)
G2_labels = data['G2'].values
G3_labels = data['G3'].values
```

Разделяем данные на train и test:

```python
from sklearn.model_selection import train_test_split
train_feature_matrix, test_feature_matrix, train_G3_labels, test_G3_labels = train_test_split(
    feature_matrix, G3_labels, test_size=0.25, random_state=42)
```

Исследуем разные методы:

```python
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import r2_score, mean_squared_error, accuracy_score
lr = LinearRegression()
lr.fit(train_feature_matrix, train_G3_labels)
pred_G3_labels = lr.predict(test_feature_matrix)
print(f'LR R^2: {r2_score(pred_G3_labels, test_G3_labels)}')
print(f'LR MSE: {mean_squared_error(pred_G3_labels, test_G3_labels)}')
for x in range(len(pred_G3_labels)):
    print(pred_G3_labels[x], test_feature_matrix[x], test_G3_labels[x])
print(lr.coef_)

rg = Ridge()
rg.fit(train_feature_matrix, train_G3_labels)
pred_G3_labels = rg.predict(test_feature_matrix)
print(f'Ridge R^2: {r2_score(pred_G3_labels, test_G3_labels)}')
print(f'Ridge MSE: {mean_squared_error(pred_G3_labels, test_G3_labels)}')    
```

```python
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
gbr = GradientBoostingRegressor()
gbr.fit(train_feature_matrix, train_G3_labels)
pred_G3_labels = gbr.predict(test_feature_matrix)
print(f'GBR R^2: {r2_score(pred_G3_labels, test_G3_labels)}')
print(f'GBR MSE: {mean_squared_error(pred_G3_labels, test_G3_labels)}')
rfr = RandomForestRegressor()
rfr.fit(train_feature_matrix, train_G3_labels)
pred_G3_labels = rfr.predict(test_feature_matrix)
print(f'RFR R^2: {r2_score(pred_G3_labels, test_G3_labels)}')
print(f'RFR MSE: {mean_squared_error(pred_G3_labels, test_G3_labels)}')
```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf.fit(train_feature_matrix, train_G3_labels)
pred_G3_labels = clf.predict(test_feature_matrix)
print(f'Tree R^2: {r2_score(pred_G3_labels, test_G3_labels)}')
print(f'Tree MSE: {mean_squared_error(pred_G3_labels, test_G3_labels)}')
print(accuracy_score(test_G3_labels, pred_G3_labels))
```
